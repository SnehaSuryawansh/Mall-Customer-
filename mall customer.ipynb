```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
dataset= pd.read_csv('Mall_Customers.csv')
dataset
```

```python
dataset.head()
```

```python
# Applying KMeans clustering algorithm
# The Elbow Method is used to find the optimal number of clusters
X = dataset.iloc[:, [3, 4]].values  # Taking Annual Income and Spending Score
wcss = []  # Within-Cluster-Sum-of-Squares
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Plotting the Elbow Method graph
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Fit the KMeans model with optimal number of clusters (say 5 from the Elbow method)
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Calculating the Silhouette Score
from sklearn.metrics import silhouette_score
silhouette_avg = silhouette_score(X, y_kmeans)
print(f'Silhouette Score: {silhouette_avg}')
```

# Hierarchial Clustering

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize= (20,6))
dendro= sch.dendrogram(sch.linkage(X,method= 'ward'))
plt.title('Dendrogram')
plt.xlabel('observations')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc= AgglomerativeClustering(n_clusters=5, linkage= 'ward')
y_hc= hc.fit_predict(X)
```

```python
plt.scatter(X[y_hc==0,0], X[y_hc==0,1], color= 'blue')
plt.scatter(X[y_hc==1,0], X[y_hc==1,1], color= 'green')
plt.scatter(X[y_hc==2,0], X[y_hc==2,1], color= 'pink')
plt.scatter(X[y_hc==3,0], X[y_hc==3,1], color= 'red')
plt.scatter(X[y_hc==4,0], X[y_hc==4,1], color= 'black')
plt.xlabel('Annual income')
plt.ylabel('Spending Score')
plt.title('Hierachial Cluster chart')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Extract relevant features (e.g., 'Annual Income (k$)' and 'Spending Score (1-100)')
X = dataset[['Annual Income (k$)', 'Spending Score (1-100)']]

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize= (20,6))
dendro= sch.dendrogram(sch.linkage(X,method= 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
# Plot clusters with labels
plt.figure(figsize=(10, 7))
plt.scatter(X_scaled[y_hc == 0, 0], X_scaled[y_hc == 0, 1], s=100, c='blue', label='Cluster 1')
plt.scatter(X_scaled[y_hc == 1, 0], X_scaled[y_hc == 1, 1], s=100, c='green', label='Cluster 2')
plt.scatter(X_scaled[y_hc == 2, 0], X_scaled[y_hc == 2, 1], s=100, c='pink', label='Cluster 3')
plt.scatter(X_scaled[y_hc == 3, 0], X_scaled[y_hc == 3, 1], s=100, c='red', label='Cluster 4')
plt.scatter(X_scaled[y_hc == 4, 0], X_scaled[y_hc == 4, 1], s=100, c='black', label='Cluster 5')

# Titles and labels
plt.title('Clusters of Customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
